# Exploration des données — Cafe Sales

## Description du dataset
Le dataset représente les ventes d’un café et contient des données transactionnelles telles que les produits vendus, les quantités, les prix, les méthodes de paiement, les lieux de vente et les dates de transaction.

## Présentation des colonnes
- Transaction ID
- Item
- Quantity
- Price Per Unit
- Total Spent
- Payment Method
- Location
- Transaction Date

## Problèmes de qualité identifiés
Lors de la première exploration du dataset, plusieurs problèmes de qualité ont été observés :
- présence de valeurs manquantes
- utilisation de valeurs génériques telles que « ERROR » et « UNKNOWN »
- incohérences dans les formats
- présence de valeurs aberrantes, notamment des prix irréalistes

Ces éléments montrent que le dataset est brut et nécessite un travail de nettoyage avant toute analyse fiable.
