# Transformation des données — Cafe Sales

## Objectif
Cette étape vise à transformer les données nettoyées afin de garantir leur cohérence et leur exploitabilité pour l’analyse.

## Recalcul et standardisation des valeurs financières
Les montants financiers ont été vérifiés afin d’assurer leur cohérence. Le montant total dépensé pour chaque transaction est cohérent avec la quantité achetée et le prix unitaire.

## Correction des valeurs aberrantes
Des valeurs aberrantes ont été identifiées, notamment des prix unitaires irréalistes. Ces valeurs ont été corrigées en utilisant la relation entre le montant total et la quantité achetée afin d’obtenir des prix unitaires réalistes et cohérents avec le contexte d’un café.

## Vérification finale
Une vérification finale a été réalisée afin de s’assurer que :
- les colonnes critiques ne contiennent plus de valeurs non exploitables
- les montants financiers sont cohérents
- les données sont prêtes pour une analyse fiable

Le dataset final est désormais propre, cohérent et exploitable.
