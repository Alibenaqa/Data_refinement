# Nettoyage des données — Cafe Sales

## Objectif
L’objectif de cette étape est de corriger les problèmes de qualité identifiés lors de l’exploration afin de rendre le dataset exploitable pour l’analyse.

## Nettoyage des colonnes critiques

### Item
Les lignes contenant des valeurs « ERROR », « UNKNOWN » ou des valeurs manquantes dans la colonne Item ont été supprimées. Le produit vendu étant une information centrale, il était impossible de corriger ces lignes de manière fiable.

### Quantity
Les lignes contenant des valeurs non numériques ou manquantes dans la colonne Quantity ont été supprimées, car la quantité est indispensable pour le calcul des montants financiers.

## Nettoyage des colonnes financières

### Total Spent
Les valeurs incorrectes ou manquantes ont été corrigées en recalculant le montant total à partir de la quantité et du prix unitaire.

### Price Per Unit
La colonne Price Per Unit contenait des valeurs aberrantes (prix irréalistes). Ces valeurs ont été corrigées en remplaçant les prix incohérents par des valeurs réalistes déduites de la relation entre le montant total et la quantité (par exemple pour le produit Tea, un prix unitaire de 1.5).

## Nettoyage des colonnes contextuelles

### Payment Method
Les valeurs « ERROR » et « UNKNOWN » ont été remplacées par des valeurs manquantes afin de conserver les transactions.

### Location
Les valeurs non exploitables ont été supprimées et remplacées par des cellules vides.

### Transaction Date
Les valeurs non valides ou manquantes ont été supprimées, la date étant une information importante pour l’analyse temporelle.
